```python
# lets import matplotlib
#use inline backend

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

# use ggplot style

mpl.style.use('classic')
```

```python
df.plot(kind='bar', x = 'month_number', y='total_profit', width = 0.81,grid=True, legend=None)

plt.xlabel('Month number')
plt.ylabel('Total Profit')
plt.title('Company profit per month', bbox ={'facecolor':'0.8','pad':2})
plt.grid(linestyle='--', linewidth = 0.2, color='black')
plt.show
```

```python
#line plot

df.plot(kind='line', x = 'month_number', y='total_profit', xlabel='Month number', ylabel='Total profit',
        title='Company profit per month', marker='o', markerfacecolor = 'black',linestyle = '--' , color = 'red',
        linewidth = '1.5')

plt.xticks(range(1,13))
plt.legend(loc='lower right')
plt.show()
```

```python
#line plot
df.plot(kind='line', x = 'month_number', y='total_profit', xlabel='Month number', ylabel='Total profit',
        title='Company profit per month', marker='o', markerfacecolor = 'black',linestyle = '--' , color = 'red',
        linewidth = '1.5', label= 'Profit data of last year')

plt.xticks(range(1,13))
plt.legend(loc='lower right')

plt.show()
```

```python
# multiline
#df.head(2)

df_product = df.iloc[:,0:-2]

df_product.plot(kind='line', x='month_number', marker='o', figsize=(10,8),xlabel='Month Number', ylabel='Sales units in number',
                title='Sales data', )

plt.xticks(range(1,13))
plt.legend(loc='upper left')
plt.show()
```

```python
#scatter plot

df.plot(kind='scatter', x = 'month_number', y = 'bathingsoap', xlabel='Month Number', ylabel='Number of unit Sold',
                title='Bathingsoap Sales Data', label= 'bathing soap Sales data')

plt.xticks(range(1,13))
plt.grid(linestyle = '--', color='gray')
plt.legend(loc='upper left')
plt.show()
```
